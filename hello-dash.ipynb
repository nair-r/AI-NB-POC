```python
import os
import ipywidgets as widgets
from IPython.display import display, HTML

display(HTML("<div class='hello-banner'>Hello from Voila &#x2714;</div>"))

env_keys = ("XNAT_HOST", "XNAT_USER", "XNAT_ITEM_ID", "XNAT_XSI_TYPE", "XNAT_DATA", "JUPYTERHUB_ROOT_DIR")
rows = "".join(
    f"<div><b>{k}</b> {os.environ.get(k, '&lt;unset&gt;')}</div>"
    for k in env_keys
)
env_html = widgets.HTML(value=f"<div class='env-table'>{rows}</div>")

click_count = {"n": 0}
out = widgets.Output()
btn = widgets.Button(description="Click me", button_style="primary")

def _on_click(_b):
    click_count["n"] += 1
    with out:
        out.clear_output()
        print(f"Clicked {click_count['n']} time{'s' if click_count['n'] != 1 else ''}.")

btn.on_click(_on_click)

display(widgets.VBox([env_html, btn, out]))
```
